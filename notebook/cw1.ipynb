```python
def resize(image, coefficient, algo):
    height, width = image.shape[:2]

    # Calculate the new dimensions
    new_height = int(height * coefficient)
    new_width = int(width * coefficient)

    # Resize the image
    resized_image = cv2.resize(image, None, fx=coefficient, fy=coefficient, interpolation=algo)

    return resized_image
```
