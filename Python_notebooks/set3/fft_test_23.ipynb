## Signal Composition and Decomposition => FFT

```python
# example of fitting a neural net on x vs x^2
%matplotlib inline 
```

```python
import tensorflow as tf
import numpy as np
from numpy import asarray
import math as m
from matplotlib import pyplot 
from scipy.fft import fft, fftfreq
```

```python
# define the dataset

x = np.arange(0,5,0.01)
y1 = asarray([ m.sin(m.pi*i)  for i in x ])
y2 = asarray([ 7*m.sin(0.5*m.pi*i-0.34)  for i in x ])
y3 = asarray([ 3*m.sin(2*m.pi*i+1.57) for i in x ])
y = y1 + y2 + y3
```

```python
pyplot.scatter(x,y, s=2, c='r', marker='.' , label='sine')
```

```python
pyplot.plot(x,y, color='red', linewidth=1)
```

```python
yf = fft(y)
print(yf.shape)
#xf = np.arange(0, yf.shape[0])
xf = fftfreq(yf.shape[0], yf.shape[0])
pyplot.plot(xf, np.abs(yf) , color='green', linewidth=1 )
```

```python
pyplot.plot(xf, np.angle(yf) , color='blue', linewidth=1 )
```

```python
# create figure
fig = pyplot.figure(figsize=(12,16))

# setting values to rows and column variables
rows = 4
columns = 2

# Adds a subplot at the 1st position
fig.add_subplot(rows, columns, 1)
# showing image
pyplot.plot(x,y1, color='red', linewidth=1)
#pyplot.axis('off')
pyplot.title("First Component y1")

fig.add_subplot(rows, columns, 3)
pyplot.plot(x,y2, color='red', linewidth=1)
#pyplot.axis('off')
pyplot.title("Second Component y2")

fig.add_subplot(rows, columns, 5)
pyplot.plot(x,y3, color='red', linewidth=1)
#pyplot.axis('off')
pyplot.title("Third Component y3")

fig.add_subplot(rows, columns, 7)
pyplot.plot(x,y, color='green', linewidth=1)
#pyplot.axis('off')
pyplot.title("Composition y")

fig.add_subplot(rows, columns, 8)
pyplot.plot(xf,np.abs(yf), color='blue', linewidth=1)
#pyplot.axis('off')
pyplot.title("its Fourier Magnitude")

```
